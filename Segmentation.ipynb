```python
import numpy as np 
import cv2
import matplotlib.pyplot as plt
import os
import argparse
```

```python
def feature_extraction(image):
    lab = cv2.cvtColor(image, cv2.COLOR_RGB2LAB)
    # set lower and upper color limits
    low_val = (0,119,128)
    high_val = (255,153,175)

    # Threshold the HSV image 
    mask = cv2.inRange(lab, low_val,high_val)
    
    # remove noise
    mask = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernel=np.ones((8,8),dtype=np.uint8))
    
    # apply mask to original image
    result = cv2.bitwise_and(image, image,mask=mask)

    #result = cv2.cvtColor(result,cv2.COLOR_HSV2RGB)
    #show image
    return image,result
```

```python
def HSV_Masking(image):
    # create hsv
     
    hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
     # set lower and upper color limits

    low_val = (0,0.137*255,0.086*255)
    high_val = (0.393*179,255,0.854*255)
    # Threshold the HSV image 
    mask = cv2.inRange(hsv, low_val,high_val)
    # remove noise
    mask = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernel=np.ones((8,8),dtype=np.uint8))
    # apply mask to original image
    result = cv2.bitwise_and(image, image,mask=mask)
    return image,result
```

```python
def segmentation(main_img):
    
    img = cv2.cvtColor(main_img, cv2.COLOR_BGR2RGB)
    gs = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
    blur = cv2.GaussianBlur(gs, (25,25),0)
    ret_otsu,im_bw_otsu = cv2.threshold(blur,0,255,cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU)
    kernel = np.ones((25,25),np.uint8)
    closing = cv2.morphologyEx(im_bw_otsu, cv2.MORPH_CLOSE, kernel)
    kernel = np.ones((50,50),np.uint8)
    closing = cv2.morphologyEx(im_bw_otsu, cv2.MORPH_CLOSE, kernel)

    output = img
    output[closing==255]=0
    return img,output
```

```python
def For_sub_directory(image_source,dest):
    folder,sub_folder = os.path.split(image_source)
    destination= os.path.join(dest,sub_folder)
    
    files=[]
    for entry in os.listdir(image_source):
        files.append(os.path.join(image_source,entry))

    for file in files:
        filename , ext = os.path.splitext(file)
        if ext.lower() != '.jpg':
            continue
            
        image = cv2.imread(file)
        image ,masked = HSV_Masking(image)
        image ,segmented = segmentation(masked)
        image ,feature = feature_extraction(segmented)
        
        folder, name = os.path.split(filename)
        new_filename = name + '_segmented'+ext
        new_filename = os.path.join(destination,new_filename)
        cv2.imwrite(new_filename,cv2.cvtColor(feature,cv2.COLOR_RGB2BGR))
        print(file,new_filename)
```
